```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
#download dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"

columns = ["age", "work-class", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital loss", "hours-per-week", "native-country", "income"]

data = pd.read_csv(url, names=columns, sep = ",", na_values = "?", skipinitialspace = True)
```

```python
data.head()
```

```python
data.describe()
```

```python
numerical_attr = data.select_dtypes(include=["int"])

plot = numerical_attr.hist(figsize=(10,12))
```

```python
categorical_attr = data.select_dtypes(include=["object"])
```

```python
for attr in ["work-class", "education", "sex", "income"]:
    fig = plt.figure(figsize=[10,3])
    sns.countplot(data = categorical_attr, x = attr).set_title(attr)
    plt.show()
```

```python
#data qual:
#check missing values
data.info()

print("Missing value count work-class: ", data['work-class'].isnull().sum())
print("Missing value count occupation: ", data['occupation'].isnull().sum())
print("Missing value count native-country: ", data["native-country"].isnull().sum())
```

```python
#drop missing values:
data = data.dropna()
data = data.drop_duplicates()
i = data[data["capital-gain"] > 80000].index
data = data.drop(i)
```

```python
#data relationships- numerical
corr = data.corr()
f, ax = plt.subplots(figsize = (10,12))
cmap = sns.diverging_palette(220, 10, as_cmap = True)
_ = sns.heatmap(corr, cmap = "YlGn", square = True, ax = ax, annot=True, linewidth = 0.1)
_.set_title('Pearson Correlation of Numerical Features', size = 15)
```

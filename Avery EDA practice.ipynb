##### I want to explore if theres a correlational relationship between any variables

```python
plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(df.corr(),annot=True)
```

### The heat map shows that theres a moderatly strong positive correlation between fixed acidity and citric acid, and also density and fixed acidity and also between the free sulpher dioxide and total sulpher dioxide

```python
sulpher_dioxide = sns.relplot(data = df, x ="free sulfur dioxide", y = "total sulfur dioxide")
```
